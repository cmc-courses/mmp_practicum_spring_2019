```python
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt

import time
from collections import defaultdict

from bch import BCH

%load_ext autoreload
%autoreload 2
```

```python
def generate_noise(n_samples, sample_size, min_errors, max_errors):
    max_errors = min(sample_size, max_errors)
    min_errors = min(max(0, min_errors), sample_size)
    noise = u = np.zeros([n_samples, sample_size], dtype=np.int)
    n_errors = np.sort(np.random.randint(min_errors, max_errors + 1, [n_samples]))
    for idx in range(n_samples):
        permutation = np.random.permutation(sample_size)
        noise[idx, permutation[:n_errors[idx]]] = 1
    return noise
```

```python
def count_statistics(n_samples, b):
    correct, mistake, reject = np.zeros([b.n + 1], dtype=np.int), np.zeros([b.n + 1], dtype=np.int), np.zeros([b.n + 1], dtype=np.int)

    for n_errors in range(0, b.n + 1):
        u = np.random.randint(0, 2, [n_samples, b.k])
        v = b.encode(u)
        noise = generate_noise(n_samples, b.n, n_errors, n_errors)
        w_noised = v ^ noise
        w_recovered = b.decode(w_noised)
        u_recovered = w_recovered[:, :b.k]

        for idx, (noise_size, n_mistakes) in enumerate(zip(np.sum(noise, axis=1), np.sum(u_recovered ^ u, axis=1))):
            if u_recovered[idx][0] < 0:
                reject[noise_size] += 1
            elif n_mistakes == 0:
                correct[noise_size] += 1
            elif n_mistakes > 0:
                mistake[noise_size] += 1
    return correct, mistake, reject
```

```python
def show_results(correct, mistake, reject, b):
    fig, ax = plt.subplots(1, 1, figsize=(15, 8))

    bar_1 = ax.bar(np.arange(0, correct.shape[0]), correct, bottom=0, label='correct decoding')
    bar_2 = ax.bar(np.arange(0, mistake.shape[0]), mistake, bottom=correct, label='decoded with mistake')    
    bar_3 = ax.bar(np.arange(0, reject.shape[0]), reject, bottom=correct + mistake, label='decoding rejection')
    ax.legend()
    ax.set_xlabel('number of changed bits', fontsize = 20)
    ax.set_ylabel('number of messages', fontsize = 20)
    
    fig.show()
```

```python
def measure_time(n_samples, b):
    n_samples, n_errors = n_samples, b.t

    u = np.random.randint(0, 2, [n_samples, b.k])
    v = b.encode(u)
    noise = generate_noise(n_samples, b.n, 0, n_errors)
    w_noised = v ^ noise

    start_time_pgz = time.time()
    w_recovered_pgz = b.decode(w_noised, method='pgz')
    end_time_pgz = time.time()

    start_time_euclid = time.time()
    w_recovered_euclid = b.decode(w_noised, method='euclid')
    end_time_euclid = time.time()

    assert np.sum(w_recovered_pgz != w_recovered_euclid) == 0
    
    u_recovered = w_recovered_pgz[:, :b.k]

    time_pgz, time_euclid = end_time_pgz - start_time_pgz, end_time_euclid - start_time_euclid

    return time_pgz, time_euclid
```

### 2.2

Проведём анализ возможностей исправления ошибок БЧХ кодом в зависимости от числа ошибок, допущенных при передаче:

```python
b = BCH(15, 3)
show_results(*count_statistics(100, b), b)
```

```python
b = BCH(15, 7)
show_results(*count_statistics(100, b), b)
```

```python
b = BCH(31, 7)
show_results(*count_statistics(100, b), b)
```

Как видно, БЧХ код действительно исправляет как минимум столько ошибок, сколько задано при его построении. Однако, при увеличении числа ошибок код становится неспособным исправлять ошибки в сообщениях. При этом, при малом числе дополнительных ошибок происходит отказ от декодирования, а при большем числе ошибок в большинстве случаев происходит ошибочное декодировние. 

### 2.3

Сравним время работы декодировщиков PGZ и Euclid:

```python
code_size, pgz, euclid = [], [], []
```

```python
for q in range(2, 8):
    n = 2 ** q - 1
    pgz_time, euclid_time = measure_time(50, BCH(n, (n - 1) // 2))
    code_size.append(n)
    pgz.append(pgz_time)
    euclid.append(euclid_time)    
```

```python
fig, ax = plt.subplots(2, 1, figsize=(15, 16))
ax[0].plot(code_size, pgz, label='pgz')
ax[0].plot(code_size, euclid, label='euclid')
ax[0].legend()
ax[0].grid(True)
ax[0].set_title('Decoding Time')
ax[0].set_xlabel('block size', fontsize = 20)
ax[0].set_ylabel('time, s.', fontsize = 20)

ax[1].plot(code_size, [_ / __ for _, __ in zip(pgz, euclid)], label='pgz/euclid')
ax[1].legend()
ax[1].grid(True)
ax[1].set_title('Time ratio')
ax[1].set_xlabel('block size', fontsize = 20)
ax[1].set_ylabel('ratio', fontsize = 20)
    
fig.show()
```

Как видно, PGZ значительно медленное чем Euclid, при этом с ростом размера блока разница между алгоритмами растёт как минимум линейно, что говорит о том, что PGZ крайне не оптимальный алгоритм.
